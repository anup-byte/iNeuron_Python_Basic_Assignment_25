Q1. What is the difference between enclosing a list comprehension in square brackets and
parentheses?

The square brackets tell Python that this is a list comprehension, producing a list. If  we use regular 
parentheses, we get a generator expression.

```python
l = [n*2 for n in range(10)] # List comprehension
g = (n*2 for n in range(10))  # Generator expression
```

```python
print(type(l))  # 'list'
print(type(g))  # 'generator'
```

Q2. What is the relationship between generators and iterators?

Iterators are objects that can be iterated on, and generators are special functions that leverage lazy 
evaluation. Implementing iterator means we must create an __iter__() and __next__() method, whereas 
a generator can be implemented using the yield keyword in a Python function or comprehension.   

Q3. What are the signs that a function is a generator function?

We can only tell it's a generator function by the presence of a yield statement. A yield statement turns a
regular function into a generator function.

Q4. What is the purpose of a yield statement?

A yield statement is similar to a return statement in its simplest form, except that instead of halting 
execution of the function and returning, yield instead delivers a value to the code looping over the generator
and stops execution of the generator function.

Q5. What is the relationship between map calls and list comprehensions? Make a comparison and
contrast between the two.

List comprehension returns a list, whereas the map function returns an object of Iterable. List comprehension
execution is faster than that of map function when the formula expression is huge and complex. Map function is
faster than list comprehension when the formula is already defined as a function earlier.

Map function:- The map function takes an Expression and an Iterable. The output will be an Iterable object where
the expression will work on each element of the given Iterable. The output of each expression will be an 
element of the resultant Iterable. 

List Comprehension:- Python List Comprehension is used for creating a list where each element is generated by 
applying a simple formula to the given list. 

Syntax and code:
List comprehension code is more concise and easier to read as compared to the map function.

map( expression, iterable) 

```python
# function to double the number
def num(n):
    return n * 2


lst = [1, 2, 3, 4, 5]

# creates a map object
x = map(num, lst)

# Print list
print(list(x))
```

List Comprehension:

[ expression for item in list if conditional ]

```python
lst = [1, 2, 3, 4, 5]

x = [i * 2 for i in lst]
print(x)
```
